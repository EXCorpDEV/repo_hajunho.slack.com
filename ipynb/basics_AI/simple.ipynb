# Jupyter notebook sample

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

### Interactive tables

Effortlessly view, navigate, sort, and filter data. Create charts and access essential data insights, including descriptive statistics and missing values – all without writing a single line of code.

```python
# Defining data for the dataframe
data = {
    'Basket': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O', 'P'],
    'Apples': [10, 20, 30, 56, 40, 40, 67, 47, 40, 4, 49, 52, 5, 56, 35, 45],
    'Bananas': [15, 6, 3, 45, 67, 44, 45, 11, 14, 18, 13, 12, 1, 34, 12, 12]
}

# Creating the dataframe
df = pd.DataFrame(data)

df
```

### Visualization in IDE

Create graphs and visualizations that match your chosen color scheme.

```python
# Calculate the sums
sum_apples = df['Apples'].sum()
sum_bananas = df['Bananas'].sum()

# Create a bar chart
plt.bar(['Apples', 'Bananas'], [sum_apples, sum_bananas], color=['red', 'blue'])

# Set a title
plt.title('Comparison of total Apples and Bananas')

# Show the plot
plt.show()
```

# **matplotlib를 사용하여 그래프 그리기**

```python
# Anaconda의 conda나 다른 환경의 Python에서 pip를 사용하여 matplotlib를 설치하려면 따옴표를 제거하세요.
'''
pip install matplotlib
#or
conda install matplotlib
'''
```

```python
import matplotlib.pyplot as plt

# 그래프(x, y)에 대한 변수 생성

x = [i for i in range(10)]
print("Values of X:",x)

y = [2 ** i for i in range(10)]
print("Values of Y:",y)
```

```python
# 여기서 해야할 일: 레이블과 제목 추가 및 셀 실행
plt.title('My First Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot(x,y)      # 기본값 linestyle='-'
```

```python
plt.xlabel('X값')
plt.ylabel('Y값')
plt.plot(x,y)
```

Matplotlib이나 IPython에서 한글 글자(예: 갑, Unicode 44050)가 현재 사용 중인 폰트에 없어서 표시할 수 없어 그래프 안의 라벨, 제목, 범례 등에 한글을 썼는데 깨져 보이거나 □(네모)로 나오는 경우가 있습니다.

macOS에서는 보통 AppleGothic, NanumGothic, Malgun Gothic 등이 기본 폰트로 사용 가능

```python
import matplotlib.font_manager as fm

# 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'

# 한글 깨짐 방지를 위해 마이너스 부호 처리
plt.rcParams['axes.unicode_minus'] = False
# 각 축에 레이블 변경하기
plt.xlabel('X값')
plt.ylabel('Y값')
plt.plot(x,y)
```

## Scatter plot

산점도(Scatter plot)는 그래프에 점을 사용하여 두 변수의 값을 표시합니다. 이를 통해 두 변수 사이에 관계 또는 상관 관계가 있는지 확인할 수 있습니다.

```python
plt.scatter(x,y)
```
