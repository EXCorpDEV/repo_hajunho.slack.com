# 회귀(Regression)

### 회귀 분석이란 무엇입니까? 회귀 분석을 하는 이유는 무엇입니까?

회귀 분석은 \*\*종속 변수(예: 판매량)\*\*와 **독립 변수(예: 광고비, 가격, 계절)** 간의 관계를 수치적으로 분석하는 기법입니다. 예측, 시계열 모델링, 그리고 변수 간 인과 관계 파악 등 다양한 목적으로 사용됩니다.

예를 들어, 한 **전자제품 회사의 마케팅 팀**이 다음 분기의 **노트북 판매량을 예측**하려고 한다고 가정해봅시다. 광고비, 제품 가격, 계절성(예: 개학 시즌), 경쟁사의 가격 전략 등 다양한 요소가 판매량에 영향을 줄 수 있습니다. 이 중에서 어떤 요소가 판매에 가장 큰 영향을 미치는지 알고 싶다면 어떻게 해야 할까요?

"광고에 더 많은 비용을 쓸수록 판매가 증가할 수 있습니다."
"가격이 저렴할수록 더 많은 소비자가 제품을 구매할 수 있습니다."

하지만 모든 요인이 유의미한 영향을 주는 건 아닐 수 있습니다. 어떤 것이 실제로 중요한지를 판단하려면?

**회귀 분석**을 통해 이를 수학적으로 검증할 수 있습니다. 회귀 분석은 마케팅 팀이 **가장 효과적인 투자 전략을 선택**하도록 도와줍니다. 예를 들어, 광고비보다 가격 전략이 더 큰 영향을 미친다면, 가격 조정이 보다 효과적인 전략일 수 있겠죠.

회귀 분석은 다음을 가능하게 합니다:

* 주요 변수 식별
* 영향력이 미미한 변수 무시
* 변수들 간 상호작용 분석

### 상관 관계 vs 인과 관계

회귀 분석에서 자주 발생하는 오류 중 하나는 **상관 관계를 인과 관계로 오해**하는 것입니다. 다음 예를 보겠습니다:

> 한 연구에서, **영화 티켓 판매량과 인터넷 검색 트래픽 증가가 높은 상관 관계**를 보였다고 합니다. 그렇다면, 사람들이 영화를 많이 볼수록 인터넷 검색을 더 한다는 뜻일까요? 혹은 인터넷 검색이 늘어날수록 영화도 더 많이 본다는 의미일까요?

사실은 두 변수 모두 **'주말'이라는 제3의 요인**과 관련이 있을 수 있습니다. 주말에는 여가 시간이 많기 때문에 영화도 보고, 검색도 많이 하게 되는 것이죠.

**상관 관계는 인과 관계를 보장하지 않습니다.**
회귀 분석을 할 때는 **도메인 지식**과 함께 결과를 해석하는 신중함이 필요합니다.
