```python
import matplotlib.pyplot as plt

#선 그래프(Line Plot)
# 데이터 설정
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 선 그래프 그리기
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()

# 선 스타일 설정하여 그래프 그리기
plt.plot(x, y, linestyle='--', color='red', marker='o')
plt.show()
```

```python
#산점도(Scatter Plot) 그리기

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 산점도 그리기
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#막대 그래프(Bar Chart)
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]

# 막대 그래프 그리기
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
#히스토그램(Histogram)

data = [0, 1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 7, 8, 8, 9, 9, 9]

# 히스토그램 그리기 (분포도)
plt.hist(data, bins=5) #각 숫자의 빈도수가 히스토그램으로 표현됩니다
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 2 * np.pi, 100)
#print('2*np.pi :', 2*np.pi)
print('Type(x) :', type(x), x.shape)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sinh(x)

print('Type(y):', type(y1), y1.shape)

# 2x2 subplot 생성
fig, axs = plt.subplots(2, 2, figsize=(10, 10)) #행2개, 열2개

# 첫 번째 플롯 - 사인 그래프
axs[0, 0].plot(x, y1, 'b')  # 'b'는 파란색을 의미
axs[0, 0].set_title('Sine Function')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('sin(x)')

# 두 번째 플롯 - 코사인 그래프
axs[0, 1].plot(x, y2, 'r')  # 'r'는 빨간색을 의미
axs[0, 1].set_title('Cosine Function')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('cos(x)')

# 세 번째 플롯 - 탄젠트 그래프
axs[1, 0].plot(x, y3, 'g')  # 'g'는 초록색을 의미
axs[1, 0].set_title('Tangent Function')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('tan(x)')
axs[1, 0].set_ylim(-5, 5)  # y축 범위 제한

# 네 번째 플롯 - 하이퍼볼릭 사인 그래프
axs[1, 1].plot(x, y4, 'y')  # 'y'는 노란색을 의미
axs[1, 1].set_title('Hyperbolic Sine Function')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('sinh(x)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
days = np.arange(1, 8)
temp_city1 = np.array([20, 21, 22, 21, 20, 19, 18])
temp_city2 = np.array([18, 17, 16, 15, 14, 13, 12])

# 두 개의 subplot 생성, y축 공유
fig, ax1 = plt.subplots(figsize=(10, 5)) # 행과 열값이 없으면 (1,1)배열. 하나의 그래프
ax2 = ax1.twinx()  # ax1과 y축을 공유하는 새로운 축 생성

# 첫 번째 도시의 온도 변화 그래프
ax1.plot(days, temp_city1, 'g-', label='City 1 Temperature')
ax1.set_xlabel('Day')
ax1.set_ylabel('City 1 Temperature', color='g')

# 두 번째 도시의 온도 변화 그래프
ax2.plot(days, temp_city2, 'b-', label='City 2 Temperature')
ax2.set_ylabel('City 2 Temperature', color='b')

# 제목 및 범례 설정
plt.title('Temperature Comparison Between Two Cities')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```
